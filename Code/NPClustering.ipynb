```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import bnpy
from sklearn.preprocessing import normalize
```

```python
data_3mer = pd.read_csv("../Data/3mermotif.tsv", delimiter="\t", index_col="icgc_sample_id")
data_3mer.shape
```

```python
data_5mer = pd.read_csv("../Data/5mermotif.tsv", delimiter="\t", index_col="icgc_sample_id")
data_5mer = data_5mer.iloc[:11474, :]
data_5mer["cancer_type"] = np.array(data_3mer["cancer_type"])
data_5mer.shape
data_5mer.to_csv("../Data/5mermotif.csv")
```

```python
# data_5mer
# data = data_5mer.drop("cancer_type", axis=1)
data = data_5mer
data.dropna(how="all", axis=1, inplace=True)
data.dropna(how="any", axis=1, inplace=True)
data = normalize(data, axis=0, norm='max')
```

```python
results = pd.read_csv("../Results/SAE.log")
results
plt.figure(figsize=(15, 10))
plt.xlabel("Epochs")
plt.ylabel("Loss")
# plt.yticks(np.arange(0, 10, 5))
plt.xticks(np.arange(0, 500, 20), rotation=90)
plt.plot(results["loss"], label="Training Loss")
plt.plot(results["val_loss"], label="Test Loss")
plt.legend(loc="best")
plt.grid()
plt.savefig("../Results/SAE.pdf")
```
